```python
import numpy as np
import time 
import matplotlib.pylab as plt
from sympy import isprime
%matplotlib inline


def FibonacciSuccession(timeInterval):
    
    '''
    Calculate the fibonacci primes in a time interval.
    input: timeInterval in seconds time.time() + seconds
    
    '''
    numbersFibonacci = [0,1]
    fibonacciPrimes = []
    while time.time() <= timeInterval:
        numbersFibonacci += [numbersFibonacci[-2] + numbersFibonacci[-1]]
        if isprime(numbersFibonacci[-1])==True:
            fibonacciPrimes += [numbersFibonacci[-1]]
        
    return fibonacciPrimes,numbersFibonacci

    
def FibonacciIndex(timeInterval):
    '''
    Return the fibonacci primes index. 
    '''
    fn,fibonacciNumbers = FibonacciSuccession(timeInterval)   
    n = np.in1d(fibonacciNumbers , fn).nonzero()[0]
    return fn,n


timeInterval = time.time() + 1e-4 # 1e-4 seconds
fn,n = FibonacciIndex(timeInterval)
print('Fibonacci Primes:',fn)
print('n: ',n)
```

<h3 align="center"> Fibonacci Primes Plot</h3>

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }


plt.figure(figsize=(9,5))
plt.title('Fibonacci Primes',fontdict=font)
plt.plot(n,fn,'k')
plt.plot(n,fn,'ro')
plt.xlabel('n', fontdict=font)
plt.ylabel('$F_n$', fontdict=font)
plt.show()
```

<h3 align="center"> In an hour </h3>

```python
timeInterval = time.time() + 60*60 # 1 hour
fn,n = FibonacciIndex(timeInterval)
print('Fibonacci Primes:',fn)
print('n: ',n)
```
