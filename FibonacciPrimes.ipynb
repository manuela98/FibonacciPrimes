```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline

def IsPrime(number):
    if number < 2:
        return False
    elif number == 2:
        return True  
    for i in range(2, number):
        if number % i ==0:
            return False
    return True

def FibonacciSuccession(N):
    '''
    Calculate the fibonacci sequence while the term is less than N
    '''
    numbersFibonacci = [0,1]
    while numbersFibonacci[-1] < N:    
        numbersFibonacci += [numbersFibonacci[-2] + numbersFibonacci[-1]]
    return numbersFibonacci

    
N = 10
fibonacciNumbers = np.array(FibonacciSuccession(N))
fibonacciPrimebool = list(map(IsPrime,fibonacciNumbers))
fn= fibonacciNumbers[fibonacciPrimebool]
n = np.in1d(fibonacciNumbers , fn).nonzero()[0]
print('Fibonacci Numbers: ',fibonacciNumbers)
print('Fibonacci Primes:',fn)
print('n: ',n)
```

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
plt.figure(figsize=(9,5))
plt.title('Fibonacci Primes',fontdict=font)
plt.plot(n,fn,'k')
plt.plot(n,fn,'ro')
plt.xlabel('n', fontdict=font)
plt.ylabel('$F_n$', fontdict=font)
plt.show()
```
